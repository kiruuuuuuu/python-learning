```python
# 11) Count Vowels and Consonants
#SOLUTION 1
def vowels_consonants(s):
    v = c = 0
    for ch in s.lower():
        if ch.isalpha():
            if ch in "aeiou":
                v += 1
            else:
                c += 1
    return v, c

print(vowels_consonants("Hello"))  # (2, 3)

#SOLUTION 2
def cv(s):
    vowel=0
    consonant=0
    for i in s:
        if i in 'aeiouAEIOU':
            vowel+=1
        else:
            consonant+=1
    print(vowel,consonant)
cv('hello')
```

```python
# 12) Count Alphabets, Digits, Special Characters
#SOLUTION 1
def count_all(s):
    a = d = sp = 0
    for ch in s:
        if ch.isalpha():
            a += 1
        elif ch.isdigit():
            d += 1
        else:
            sp += 1
    return a, d, sp

print(count_all("abc123!@"))  # (3, 3, 2)

#SOLUTION 2
def all(s):
    a=d=ch=0
    for i in s:
        if (i>='a' and i<='z') or (i>='A' and i<='Z'):
            a+=1
        elif i>='0' and i<='9':
            d+=1
        else:
            ch+=1
    print(a,d,ch)
all('abc123!@')
```

```python
# 1) Access List Index and Values

def access_list(lst):
    result = []
    for i in range(len(lst)):
        result.append((i, lst[i]))
    return result

print(access_list([10, 20, 30]))
```

```python
# 2) Add Two Lists
def add_lists(a, b):
    result = []
    for i in range(len(a)):
        result.append(a[i] + b[i])
    return result

print(add_lists([1,2], [3,4]))  # [4,6]
```

```python
# 3) Arithmetic Operations
def list_arithmetic(a, b):
    add = []
    sub = []
    mul = []
    for i in range(len(a)):
        add.append(a[i] + b[i])
        sub.append(a[i] - b[i])
        mul.append(a[i] * b[i])
    return add, sub, mul

print(list_arithmetic([4,5], [1,2]))
```

```python
# 4) Check List Empty
def is_empty(lst):
    count = 0
    for i in lst:  #can use _ instead of i
        count += 1
    return count == 0

print(is_empty([]))  # True
```

```python
# 8) Largest Number
def largest(lst):
    max_val = lst[0]
    for x in lst:
        if x > max_val:
            max_val = x
    return max_val

print(largest([1,9,3]))  # 9
```

```python
# 9) Largest and Smallest
def large_small(lst):
    max_val = lst[0]
    min_val = lst[0]
    for x in lst:
        if x > max_val:
            max_val = x
        if x < min_val:
            min_val = x
    return max_val, min_val

print(large_small([1,9,3]))
```

```python
# 11) List Difference       
def list_diff(a, b):
    result = []
    for x in a:
        found = False
        for y in b:
            if x == y:
                found = True
                break
        if not found:
            result.append(x)
    return result

print(list_diff([1,2,3], [2,4,9]))




def list_diff(a, b):
    result = []
    li = a+b
    
    for x in li:   # combine both lists
        found = False
        
        if x in a and x in b:
            found = True   # skip common elements
        
        if not found:
            result.append(x)
    
    return result

print(list_diff([1,2,3], [2,4,9]))
```

```python
# 20) Remove Duplicates
def remove_duplicates(lst):
    result = []
    for x in lst:
        duplicate = False
        for y in result:
            if x == y:
                duplicate = True
                break
        if not duplicate:
            result.append(x)
    return result

print(remove_duplicates([1,2,2,3]))
```

```python
# 22) Reverse List
def reverse_list(lst):
    result = []
    i = 0
    while i < len(lst):
        result = [lst[i]] + result
        i += 1
    return result

print(reverse_list([1,2,3]))
```

```python
# 23) Second Largest        
def second_largest(lst):
    first = second = float('-inf')
    for x in lst:
        if x > first:
            second = first
            first = x
        elif x > second and x != first:
            second = x
    return second

print(second_largest([1,5,3,9]))
```

```python
# 24) Sort Ascending
def sort_list(lst):
    n = len(lst)
    for i in range(n):
        for j in range(0, n-i-1):
            if lst[j] > lst[j+1]:
                lst[j], lst[j+1] = lst[j+1], lst[j]
    return lst

print(sort_list([3,1,2]))
```

```python
# 25) Smallest Element
def smallest(lst):
    min_val = lst[0]
    for x in lst:
        if x < min_val:
            min_val = x
    return min_val

print(smallest([3,1,2]))
```

```python
# 1. Reverse Number
def reverse_number(n):
    rev = 0
    while n > 0:
        rev = rev * 10 + n % 10
        n = n // 10
    return rev

num = int(input("Enter a number to reverse: "))
print("Reversed number:", reverse_number(num))
```

```python
# 2. Find Sum of Digits
def sum_of_digits(n):
    total = 0
    while n > 0:
        total += n % 10
        n = n // 10
    return total

num = int(input("Enter a number to find sum of digits: "))
print("Sum of digits:", sum_of_digits(num))
```

```python
# 3. Check Palindrome
def reverse_number(n):
    rev = 0
    while n > 0:
        rem = n % 10
        rev = rev * 10 + rem
        n = n // 10
    return rev

def is_palindrome(n):
    return n == reverse_number(n)

num = int(input("Enter a number to check palindrome: "))
if is_palindrome(num):
    print(num, "is a palindrome")
else:
    print(num, "is not a palindrome")
```

```python
# 4. Find Largest of Two Numbers
def largest_of_two(a, b):
    if a > b:
        return a
    else:
        return b

a = int(input("Enter first number: "))
b = int(input("Enter second number: "))
print("Largest number is:", largest_of_two(a, b))
```

```python
# 5. Check Armstrong Number
def is_armstrong(n):
    temp = n
    sum = 0
    digits = 0
    t = n
    while t > 0:
        digits += 1
        t = t // 10
    while temp > 0:
        r = temp % 10
        sum += r ** digits
        temp = temp // 10
    return sum == n

num = int(input("Enter a number to check Armstrong: "))
if is_armstrong(num):
    print(num, "is an Armstrong number")
else:
    print(num, "is not an Armstrong number")
```

```python
# 6. Check Strong Number
def factorial(n):
    fact = 1
    for i in range(1, n+1):
        fact *= i
    return fact

def is_strong(n):
    temp = n
    sum = 0
    while temp > 0:
        sum += factorial(temp % 10)
        temp = temp // 10
    return sum == n
num = int(input("Enter a number to check Strong number: "))
if is_strong(num):
    print(num, "is a Strong number")
else:
    print(num, "is not a Strong number")
```

```python
# 7. Check Perfect Number
def is_perfect(n):
    sum = 0
    for i in range(1, n):
        if n % i == 0:
            sum += i
    return sum == n

num = int(input("Enter a number to check Perfect number: "))
if is_perfect(num):
    print(num, "is a Perfect number")
else:
    print(num, "is not a Perfect number")
```

```python
# 8. Fibonacci Series       
def fibonacci(n):
    a, b = 0, 1
    fib_series = []
    for _ in range(n):
        fib_series.append(a)
        a, b = b, a + b
    return fib_series

n = int(input("Enter number of terms for Fibonacci: ")) #6
print("Fibonacci series:", fibonacci(n))
```

```python
# 9. Count of Digits
def count_digits(n):
    count = 0
    if n == 0:
        return 1
    while n > 0:
        count += 1
        n = n // 10
    return count

num = int(input("Enter a number to count digits: "))
print("Number of digits:", count_digits(num))
```

```python
# 10. Sum of First n Natural Numbers
def sum_natural(n):
    total = 0
    for i in range(1, n+1):
        total += i
    return total

n = int(input("Enter number to sum first n natural numbers: "))
print("Sum of first", n, "natural numbers:", sum_natural(n))
```

```python
# 11. Power of Number
def power(a, b):
    result = 1
    for _ in range(b):
        result *= a
    return result

base = int(input("Enter base: "))
exp = int(input("Enter exponent: "))
print(f"{base}^{exp} =", power(base, exp))
```

```python
# 12. Check Neon Number
def is_neon(n):
    sq = n * n
    sum_digits = 0
    while sq > 0:
        sum_digits += sq % 10
        sq = sq // 10
    return sum_digits == n

num = int(input("Enter a number to check Neon: "))
if is_neon(num):
    print(num, "is a Neon number")
else:
    print(num, "is not a Neon number")
```

```python
# 13. Check Spy Number
def is_spy(n):
    sum_digits = 0
    product_digits = 1
    temp = n
    while temp > 0:
        r = temp % 10
        sum_digits += r
        product_digits *= r
        temp = temp // 10
    return sum_digits == product_digits

num = int(input("Enter a number to check Spy number: "))
if is_spy(num):
    print(num, "is a Spy number")
else:
    print(num, "is not a Spy number")
```
